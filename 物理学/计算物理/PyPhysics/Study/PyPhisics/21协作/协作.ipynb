# 21 协作

## 21.1 任务单系统（Issue ticketing System）

简化协作方案：
任务管理系统（ticketing System）/内容管理系统（content management system）/问题跟踪器（issue tracker）

基于Web的任务单管理系统允许为任务分配、更新和完成报告提供Web界面，在单个任务级别中跟踪和管理项目。
基于Web的代码库托管服务几乎都有问题跟踪器，可以创建与代码库相关的问题或任务单。
生成的记录板是一个动态说明系统，用于待办事项列表管理、通信、同行评审等。

### 21.1.2 创建问题

GitHub上的“问题（Issue）”是与特定代码库关联的任务单。
问题向代码开发人员和用户报告错误、请求功能并告知已知错误。

任务单专门指定了一个用于讨论并更新相关主题的地方。

#### 问题的核心特性

问题要包含的几个关键问题：
1.需要做些什么？即描述错误或请求的功能。
2.为什么要这么做？提供足够的信息来复现错误。
3.接下来发生什么？关于进一步工作的提议。
4.如何衡量这个问题会得到解决？明确最终目标。

#### 问题元数据

除了核心信息，还可以向问题添加一些有助于组织问题并指出问题的位置的元数据。
1.不同的基于Web的托管平台提供了不同的特性来定义任务单。
2.GitHub中的问题可以使用一些整洁的特性，包括标签、通知某个用户、添加与代码库的交叉链接，以及添加提交链接等。
3.每个项目可完全定义自己的标签，用于对问题进行分类和区分。

单个问题可以使用多个标签。
在GitHub上还可以使用颜色用于帮助在视觉上区分不同的主题。

问题标签可用于以不同的方式对问题进行分类，可以指示重要性的程度、复杂的程度、问题的类型、受影响的代码的组件或甚至问题的状态。

### 21.1.3 分配问题

将问题分配给特定的开发人员：避免重复的工作。
问题可以保留未分配的状态，以表示无人认领。此时对解决这些问题感兴趣的开发者可以将问题分配给自己。

Github里程碑：一组问题，定义了更广泛的目标，有截止日期。可以用作许多类型的项目目标跟踪的机制，包括研究组组织和代码发布管理。

### 21.1.4 讨论问题

在GitHub上讨论一个问题，只需在与问题相关的评论框中输入一条新评论。

问题会话目的：
1.询问和回答有关的澄清问题。
2.分享和讨论解决方案的想法。
3.请求并提供过程状态的更新。

电子邮件的形式通常更适合开放讨论，且没有时间限制。

### 21.1.5 关闭问题

当修复错误或实现新功能时，就应该关闭该问题。
协作的文化规范、专业知识分布、领导层次、验证和审核需求都会影响到问题完成的过程。当问题不再违反其要求时，问题可以被关闭。

科学界的不同：
希望科学软件具有高水平的稳健性和质量。
为了确保质量和准确性，对科学软件项目的新更改可能需要经过审核、验证、同行评审。
**此时关闭问题可能要满足多个条件。**

## 21.2 拉取请求和代码审查

电子邮件传递补丁共享、删除和提交更改$\longrightarrow\limits^{演化为}$拉取请求

拉取请求：一个合理的、可知道源谱的接口，用于对提交的补丁进行同行审查。

### 21.2.1 提交拉取请求

拉取请求接口使得研究人员可以在一个干净的界面中提交一个修改，以链接到实际的提交。
允许对代码进行评论，并在GitHub服务器上持久化。

拉取请求可以在Github上进行。

### 21.2.2 审查合并请求

审阅拉取请求审阅代码中的一段。
开发人员应该尽可能避免冗长或复杂的请求。
通过以原子方式解决变化（一次修复一个缺陷或增加功能），开发人员可以减少引入可能在审查阶段遗漏的错误的可能性。

审查拉取请求的开发者询问代码是否：
1.实现了目标？
2.是否引入了错误？
3.包含足够的测试？
4.遵循样式指南？
5.通过了现有测试？
6.通过了新测试？
7.在其他平台（UNIX、Windows）上通过了测试？

### 21.2.3 合并拉取请求

审查结束后，就可以合并代码。

两种方式：
1.在GitHub 上，拉取请求本身有一个绿色按钮用于合并不冲突的拉取请求。
2.通过命令行，使用git remote、git fetch、git merge、git push的组合。
